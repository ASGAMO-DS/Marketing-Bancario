```python
#Factorizando las variables categóricas

df['job_num']= pd.factorize(df['job'])[0]
df['marital_num']= pd.factorize(df['marital'])[0]
df['education_num']= pd.factorize(df['education'])[0]
df['default_num']= pd.factorize(df['default'])[0]
df['housing_num']= pd.factorize(df['housing'])[0]
df['loan_num']= pd.factorize(df['loan'])[0]
df['contact_num']= pd.factorize(df['contact'])[0]
df['month_num']= pd.factorize(df['month'])[0]
df['day_of_week_num']= pd.factorize(df['day_of_week'])[0]
df['poutcome_num']= pd.factorize(df['poutcome'])[0]
df['y_num']= pd.factorize(df['y'])[0]
```

```python
#Análisis Multivariante
#Análisis numérico - numérico (¿o esto sería numérico categórico porque se factorizó?)

fig, axis = plt.subplots(10, 2, figsize = (7,25))

sns.regplot(ax = axis[0,0], data=df,x='age', y='y_num')
sns.heatmap(df[['y_num','age']].corr(),annot=True,fmt='.2f',ax=axis[1,0],cbar=False)

sns.regplot(ax = axis[0,1], data=df,x='duration', y='y_num')
sns.heatmap(df[['y_num','duration']].corr(),annot=True,fmt='.2f',ax=axis[1,1],cbar=False)

sns.regplot(ax = axis[2,0], data=df,x='campaign', y='y_num')
sns.heatmap(df[['y_num','campaign']].corr(),annot=True,fmt='.2f',ax=axis[3,0],cbar=False)

sns.regplot(ax = axis[2,1], data=df,x='pdays', y='y_num')
sns.heatmap(df[['y_num','pdays']].corr(),annot=True,fmt='.2f',ax=axis[3,1],cbar=False)

sns.regplot(ax = axis[4,0], data=df,x='previous', y='y_num')
sns.heatmap(df[['y_num','previous']].corr(),annot=True,fmt='.2f',ax=axis[5,0],cbar=False)

sns.regplot(ax = axis[4,1], data=df,x='emp.var.rate', y='y_num')
sns.heatmap(df[['y_num','emp.var.rate']].corr(),annot=True,fmt='.2f',ax=axis[5,1],cbar=False)

sns.regplot(ax = axis[6,0], data=df,x='cons.price.idx', y='y_num')
sns.heatmap(df[['y_num','cons.price.idx']].corr(),annot=True,fmt='.2f',ax=axis[7,0],cbar=False)

sns.regplot(ax = axis[6,1], data=df,x='cons.conf.idx', y='y_num')
sns.heatmap(df[['y_num','cons.conf.idx']].corr(),annot=True,fmt='.2f',ax=axis[7,1],cbar=False)

sns.regplot(ax = axis[8,0], data=df,x='euribor3m', y='y_num')
sns.heatmap(df[['y_num','euribor3m']].corr(),annot=True,fmt='.2f',ax=axis[9,0],cbar=False)

sns.regplot(ax = axis[8,1], data=df,x='nr.employed', y='y_num')
sns.heatmap(df[['y_num','nr.employed']].corr(),annot=True,fmt='.2f',ax=axis[9,1],cbar=False)

plt.tight_layout()
plt.show()
```
